# Test for Diffractio() class

```python
from diffractio import np
from diffractio import um, degrees

from diffractio.diffractio import Diffractio
```

```python
x = np.linspace(-10 * um, 10 * um, 100)
y = np.linspace(-10 * um, 10 * um, 100)
z = np.linspace(-10 * um, 10 * um, 100)

x_none = None
y_none = None
z_none = None

wavelength = 0.6328 * um
info = "Test"
```

```python
keys = ['kind', 'frame', 'x', 'y', 'z', 'wavelength', 'n_background', 'info']
```

```python
list_options = []
list_options.append(['scalar', 'field', x, y_none, z_none, wavelength, 1., 'scalar_field_x'])
list_options.append(['scalar', 'field', x, y, z_none, wavelength, 1.,'scalar_field_xy'])
list_options.append(['scalar', 'field', x, y_none, z, wavelength, 1.,'scalar_field_xz'])
list_options.append(['scalar', 'field', x, y, z, wavelength, 1.,'scalar_field_xyz'])   
list_options.append(['scalar', 'field', x_none, y_none, z, wavelength, 1. ,'scalar_field_z'])   
list_options.append(['scalar', 'mask', x, y_none, z_none, wavelength, 1. ,'scalar_mask_x']) 
list_options.append(['scalar', 'mask', x, y, z_none, wavelength, 1. ,'scalar_mask_xy'])
list_options.append(['scalar', 'mask', x, y_none, z, wavelength, 1. ,'scalar_mask_xz'])
list_options.append(['scalar', 'mask', x, y, z, wavelength, 1. ,'scalar_mask_xyz'])
list_options.append(['scalar', 'source', x, y_none, z_none, wavelength, 1. ,'scalar_source_x'])
list_options.append(['scalar', 'source', x, y, z_none, wavelength, 1. ,'scalar_source_xy'])
list_options.append(['scalar', 'field', x, y, z, wavelength, 1. ,'scalar_field_xyz'])
list_options.append(['vector', 'field', x, y_none, z_none, wavelength, 1. ,'vector_field_x'])
list_options.append(['vector', 'field', x, y, z_none, wavelength, 1. ,'vector_field_xy'])
list_options.append(['vector', 'mask', x, y, z_none, wavelength, 1. ,'vector_mask_xy'])
list_options.append(['vector', 'source', x, y, z_none, wavelength, 1. ,'vector_source_xy'])
list_options.append(['vector', 'field', x, y_none, z, wavelength, 1. ,'vector_field_xz'])
list_options.append(['vector', 'field', x, y, z, wavelength, 1. ,'vector_field_xyz'])
list_options.append(['vector', 'field', x_none, y_none, z, wavelength, 1. ,'vector_field_z'])
```

```python
new_dict = {}
for i, lista in enumerate(list_options):
    print("{}/{}".format(i,lista[-1]))
    for key, value in zip(keys, lista):
        new_dict[key] = value
    u = Diffractio(**new_dict)
    print(u.__class__)
    print(u)
    print("_________________\n")
```
